### 5.3 Imputation Strategies for the remaining Columns

Looking at the remaining Columns, Some of these indicators are slow-changing socioeconomic and energy-related variables that typically evolve gradually over time rather than fluctuating abruptly from year to year. Therefore, a country’s most recent available value is often a reasonable approximation for a nearby missing year.  
Therefore, for slow-changing indicators like `elec_access_pct`, `gdp_per_capita_usd`, `co2_kt`, `renew_share_pct`, etc. we will use forward-fill (ffill) within each country. 

This approach will fill missing entries using the immediately preceding available value from the same country, preserving country-specific temporal patterns and ensuring smoother longitudinal continuity in the dataset.





For electricity generation columns like `elec_fossil_TWh`, `elec_renew_TWh`, `elec_nuclear_TWh`. These variables represent measured electricity generation outputs by different energy sources. In many cases, missing values in these columns likely indicates that a country did not generate electricity from a particular source during a given year rather than a failure in data collection, especially in terms of nuclear electricity generation in countries without nuclear infrastructure or renewable generation in countries with limited renewable capacity.

Therefore, interpreting these missing values as zero generation is more meaningful than treating them as unknown values. Hence, we'll fill missing values with a `0`. We assume that no electricity generation was recorded from the corresponding energy source for that country and year. This approach preserves the interpretability of the energy generation indicators while maintaining consistency across the dataset.



**Note:** We use Median over Mean here because these data can be skewed by outlier years (e.g., a drought hitting hydropower), So the median is a more robust representation of a country's typical energy state than the mean.

```python
# Grouping by Entity (Country) is CRITICAL to ensure data doesn't "leak" between nations
# ensure data arrays are structurally stable, chronologically sequenced
africa_df = africa_df.sort_values(['country', 'year']).reset_index(drop=True)

# 1. Forward-fill slow-changing indicators within each country
slow_indicators = [
    'elec_access_pct','clean_fuels_pct','renew_share_pct',
              'renew_cap_per_capita_W','energy_intensity_MJ_per_USD',
              'gdp_growth_pct','gdp_per_capita_usd', 'co2_Mt'
]
for col in slow_indicators:
    africa_df[col] = africa_df.groupby('country')[col].ffill()

# 2. Fill electricity generation columns with 0
gen_columns = ['elec_fossil_TWh','elec_nuclear_TWh','elec_renew_TWh','elec_total_TWh']
africa_df[gen_columns] = africa_df[gen_columns].fillna(0)

# output to show only the Columns that still contain missing values, along with their exact null counts.
print('Columns still containing missing values & their exact null counts:')
print(africa_df[slow_indicators + gen_columns].isnull().sum()[lambda s: s > 0])
```

Despite our cleaning procedures, we still have some columns with remaining missing values.

we shall now loop through the columns and print a summary showing the Countries with these missing values and the year range (Min to Max) for those gaps.


```python
cols_to_check = ['elec_access_pct', 'clean_fuels_pct', 'gdp_growth_pct', 'gdp_per_capita_usd', 'co2_Mt']

for col in cols_to_check:
    if col in africa_df.columns:
        print(f"\n--- [ {col} ] Missing Value Breakdown ---")
        # Filter for rows where this specific column is null
        missing = africa_df[africa_df[col].isnull()]
        
        if missing.empty:
            print("No missing values found.")
        else:
            # Group by Country to see the 'shape' of the missingness
            breakdown = missing.groupby('country')['year'].agg(['count', 'min', 'max'])
            print(breakdown)
    else:
        print(f"\nColumn '{col}' not found in DataFrame. Please check your column names.")
```

From the results these remaining gaps fall into different categories and will require different fixes.



***elec_access_pct:*** The remaining missing values in `elec_access_pct` were concentrated in the early years for Guinea-Bissau and Liberia.  
Electricity access is a slow-changing infrastructure indicator that generally evolves gradually over time. Since the missing values occurred at the beginning of the time series, backward propagation from subsequent years provides a realistic approximation.
Missing values in `elec_access_pct` were filled using backward-fill (bfill) within each country group, allowing later observed values to estimate earlier missing years while preserving country-specific trends.


***gdp_per_capita_usd:*** Some countries, including Congo, Egypt, and Gambia, contained extended missing periods in `gdp_per_capita_usd`.  
GDP per capita is a core economic indicator and widely considered important for sustainability analysis. Since full country-level historical data was unavailable in some cases, regional-year medians provided a more realistic approximation than global averages by preserving both temporal and regional economic context.  
Missing values in `gdp_per_capita_usd` shall be imputed using the median GDP per capita of countries within the same African region and year. This approach maintained regional economic consistency while reducing the influence of extreme values.


***co2_Mt:*** The `co2_Mt` variable contained complete missing series for a few countries, including Egypt and Gambia. This CO₂ emissions are closely related to regional economic activity, industrialization, and energy consumption patterns.  
Since country-level data are unavailable for the entire period, regional year estimates can provide a practical approximation while preserving temporal and geographic context. So we shall fill its Missing values using regional year median values.  
These imputed observations will be treated as estimated values rather than exact measurements and will be interpreted cautiously during analysis.


***clean_fuels_pct:*** Libya contained missing values for all 20 years of the study period in the `clean_fuels_pct` variable.  Because no country-specific historical values existed, our country-level imputation methods could not work. However, Libya shares similar economic and energy characteristics with other North African countries, making regional estimates a reasonable approximation.
Therefore, we shall fill its missing values using the yearly regional mean for North African countries in order to preserve regional energy access characteristics and maintain dataset completeness.


***gdp_growth_pct:*** The `gdp_growth_pct` variable contains substantial missingness across multiple countries and years, including several complete country-level gaps. Since `gdp_growth_pct` is a highly volatile macroeconomic indicator that can fluctuate sharply due to political instability, recessions, commodity-price shocks, and external economic conditions. Reliable imputation would therefore be difficult and could introduce unrealistic economic trends into the dataset. Since the variable is not central to the project’s sustainability analyses.
We will drop the `gdp_growth_pct` variable from the dataset to avoid introducing unreliable synthetic estimates.

```python
### Note: Libya
```

```python
# 1. bfill early-series access gaps
# (likely drop) africa_df['elec_access_pct'] = africa_df.groupby('country')['elec_access_pct'].bfill()

africa_df['elec_access_pct'] = (africa_df.groupby('country')['elec_access_pct']
    .transform(lambda x: x.bfill()))


# 2. Libya clean fuels — regional mean
africa_df['clean_fuels_pct'] = africa_df.groupby(['sub_region', 'year'])['clean_fuels_pct']\
    .transform(lambda x: x.fillna(x.mean()))

# 3. GDP growth — too volatile to impute full series, drop
africa_df.drop(columns=['gdp_growth_pct'], inplace=True)

# 4. Flag then fill GDP per capita
africa_df['gdp_imputed'] = africa_df['gdp_per_capita_usd'].isna()
africa_df['gdp_per_capita_usd'] = africa_df.groupby(['sub_region', 'year'])['gdp_per_capita_usd']\
    .transform(lambda x: x.fillna(x.median()))

# 5. CO2 — regional-year median
africa_df['co2_imputed'] = africa_df['co2_Mt'].isna() # Flag
africa_df['co2_Mt'] = africa_df.groupby(['sub_region', 'year'])['co2_Mt']\
    .transform(lambda x: x.fillna(x.median()))

# Verify remaining missing values
remaining = africa_df.isnull().sum()

if (remaining > 0).any():
    print(remaining[remaining > 0])
else:
    print("No remaining missing values.")
```

Variables with excessive or unreliable missingness, such as `renew_equiv_pct` and `gdp_growth_pct`, were removed from the dataset because robust imputation would likely introduce unrealistic values and distort future analysis.

Different imputation strategies were applied depending on the nature of each variable. Slow-changing indicators such as `elec_access_pct` were filled using backward-fill within each country to preserve temporal continuity. Electricity-generation variables and `financial_flows_bn_usd` were filled with 0 under the assumption that missing values represented no recorded activity rather than missing measurement.

For variables with partial but meaningful regional structure, regional-year imputation methods were used. Missing values in `gdp_per_capita_usd` and `co2_Mt` were filled using sub-regional yearly median values, while Libya’s missing `clean_fuels_pct` values were filled using the yearly North African regional mean. These approaches preserved regional economic and energy characteristics while minimizing distortion from global averages.

To improve transparency and future analytical reliability, imputation flags such as `gdp_imputed` and `co2_imputed` were created before filling missing values. These indicators identify records that were statistically estimated rather than directly observed, allowing future analyses and predictive models to account for potential uncertainty associated with imputed observations.

Overall, the cleaning workflow combined variable-specific imputation strategies, contextual interpretation of missingness, and selective feature removal to produce a more reliable and analytically consistent dataset suitable for exploratory analysis, visualization, and predictive modelling of sustainable energy trends across African countries.

```python
# Calculate shares in elec_total_TWh 
region_year['fossil_share_pct'] = (region_year['elec_fossil_TWh'] / region_year['elec_total_TWh']) * 100
region_year['nuclear_share_pct'] = (region_year['elec_nuclear_TWh'] / region_year['elec_total_TWh']) * 100
region_year['renew_share_pct'] = (region_year['elec_renew_TWh'] / region_year['elec_total_TWh']) * 100

# Plot each sub-region
fig, axes = plt.subplots(2, 3, figsize=(14, 7), sharex=True, sharey=True)
axes = axes.flatten()

for i, ar in enumerate(region_year['sub_region'].unique()):

    reg = region_year[region_year['sub_region'] == ar]

    axes[i].stackplot(
        reg['year'],
        reg['fossil_share_pct'],
        reg['nuclear_share_pct'],
        reg['renew_share_pct'],
        labels=['Fossil', 'Nuclear', 'Renewables'],
        alpha=0.8
    )

    axes[i].set_title(ar, fontsize=10)
    axes[i].set_xticks(range(2000, 2021, 5))
    axes[i].set_ylim(0, 100)

# Common labels
fig.supxlabel('Year')
fig.supylabel('Share of Total Electricity (%)')
fig.suptitle(
    'Electricity Generation Mix by African Sub-Region (2000–2019)\n'
    '(Share of Fossil, Nuclear, and Renewable Electricity)',
    fontsize=14,
    fontweight='bold'
)


# One shared legend
fig.legend(['Fossil', 'Nuclear', 'Renewables'], loc='upper right')
plt.tight_layout(rect=[0, 0, 1, 0.94])

plt.tight_layout()
plt.savefig(V_EM / 'em1_regional_generation_mix.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 9.3 Renewable Energy Transition Performance Across African Countries


* To determine which African countries made the greatest and least progress in integrating renewable energy infrastructure between 2000 and 2019?

```python
# Select relevant columns
renew_trend = clean_df[['country', 'year', 'elec_renew_TWh', 'elec_total_TWh']].dropna()

# Remove zero values to avoid division errors
renew_trend = renew_trend[renew_trend['elec_total_TWh'] > 0]

# Calculate renewable share for each year
renew_trend['renew_share_pct'] = (
    renew_trend['elec_renew_TWh'] / renew_trend['elec_total_TWh']
) * 100

# Extract start and end years
start = renew_trend[renew_trend['year'] == 2000][['country', 'renew_share_pct']]
start = start.rename(columns={'renew_share_pct': 'renew_2000'})

end = renew_trend[renew_trend['year'] == 2019][['country', 'renew_share_pct']]
end = end.rename(columns={'renew_share_pct': 'renew_2019'})

# Merge and calculate improvement
improve = start.merge(end, on='country', how='inner')
improve['renewable_improvement'] = improve['renew_2019'] - improve['renew_2000']

# Top 10 and Bottom 10
top10 = improve.sort_values('renewable_improvement', ascending=False).head(10)
bottom10 = improve.sort_values('renewable_improvement', ascending=True).head(10)

# Create subplots (same frame)
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharex=False)

# ---- Left plot: Top performers ----
axes[0].barh(top10['country'], top10['renewable_improvement'],
             color='#2A9D8F', alpha=0.9)

for i, val in enumerate(top10['renewable_improvement']):
    axes[0].text(val + 1, i, f'{val:.1f}%', va='center', fontsize=8)

axes[0].set_title('Top 10 Renewable Integration Improvers')
axes[0].set_xlabel('Change in Renewable Share (%)')
axes[0].invert_yaxis()   # Highest at top
axes[0].grid(axis='x', alpha=0.2)

# ---- Right plot: Bottom performers ----
axes[1].barh(bottom10['country'],
             bottom10['renewable_improvement'],
             color='#E63946', alpha=0.9)

for i, val in enumerate(bottom10['renewable_improvement']):
    axes[1].text(val - 5, i, f'{val:.1f}%', va='center', fontsize=8)

axes[1].set_title('Bottom 10 Renewable Integration Performers')
axes[1].set_xlabel('Change in Renewable Share (%)')
axes[1].invert_yaxis()
axes[1].grid(axis='x', alpha=0.2)

# Main title
fig.suptitle('Renewable Energy Integration Performance in Africa (2000–2019)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(V_RT / 'rt3_renew_integration_performance.png', dpi=150, bbox_inches='tight')

plt.show()
```

###  10.2 Financial Flows Vs Renewable Energy Share Growth in Africa


```python
# Select required columns
fin = clean_df[['country', 'year', 'financial_flows_bn_usd', 'renew_share_pct',
    'renew_cap_per_capita_W', 'sub_region']].dropna()

# First available renewable share
start = fin.sort_values('year').groupby('country', as_index=False) \
    .first()[['country', 'renew_share_pct']] \
    .rename(columns={'renew_share_pct': 'renew_start'})

# Last available renewable share
end = fin.sort_values('year').groupby('country', as_index=False) \
    .last()[['country', 'renew_share_pct', 'renew_cap_per_capita_W', 'sub_region']] \
    .rename(columns={'renew_share_pct': 'renew_end'})

# Total financial flows across all years
flows = fin.groupby('country', as_index=False)['financial_flows_bn_usd'].sum()

# Convert Billion USD → Million USD
flows['financial_flows_musd'] = ( flows['financial_flows_bn_usd'] * 1000)

# Merge datasets
growth = start.merge(end, on='country', how='inner')
growth = growth.merge( flows[['country', 'financial_flows_musd']], on='country', how='inner')

# Total renewable share growth
growth['renew_growth_pct'] = (growth['renew_end'] - growth['renew_start'])

# Remove invalid values
growth = growth[growth['financial_flows_musd'] > 0]

# Plot
fig, ax = plt.subplots(figsize=(14, 8))

# Scatter by region
for region, grp in growth.groupby('sub_region'):
    ax.scatter( grp['financial_flows_musd'], grp['renew_growth_pct'], s=70,
        color=REGION_COLORS.get(region, 'grey'),
        alpha=0.75, edgecolors='white', linewidth=0.7, label=region)

    # Country labels
    for _, row in grp.iterrows():
        ax.annotate( row['country'], (row['financial_flows_musd'],
         row['renew_growth_pct']), fontsize=7, alpha=0.8, xytext=(3, 2), textcoords='offset points')

# Regression
x_ = np.log10(growth['financial_flows_musd'].values)
y_ = growth['renew_growth_pct'].values
slope, intercept, r, p, _ = stats.linregress(x_, y_)
xfit = np.linspace(x_.min(), x_.max(), 200)
yfit = slope * xfit + intercept
ax.plot( 10**xfit, yfit, 'k--', lw=1.5, alpha=0.7,
    label=f'Linear fit  r={r:.2f}  p={p:.3f}')
ax.plot(10**xfit,slope*xfit+intercept,'k--',lw=1.5,alpha=0.6,label=f'r={r:.2f}  p={p:.3f}')


# Median reference lines
median_fin = growth['financial_flows_musd'].median()
median_growth = growth['renew_growth_pct'].median()
ax.axvline(median_fin, color='grey', linestyle=':', alpha=0.7)
ax.axhline(median_growth, color='steelblue', linestyle=':', alpha=0.7)

# Axis scaling
ax.set_xscale('log')

# Custom x-axis ticks
ax.set_xticks([1, 10, 50, 100, 500, 1000, 5000, 10000])
ax.set_xticklabels(
    ['1', '10', '50', '100', '500', '1000', '5000', '10000'])

# Labels and title
ax.set_xlabel('Total International Financial Inflow (Million USD, log scale)',fontsize=11)
ax.set_ylabel('Absolute Renewable Share Growth (2000 → 2019, pp)', fontsize=11)
ax.set_title('International Financial Flows vs Renewable Energy Transition (2000–2019)', 
             fontsize=13, fontweight='bold')

ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig(V_RT / 'rt5_financialflow_vs_transition.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Pearson r = {r:.3f}  |  p-value = {p:.4f}')
```

## 12. CO₂ Emissions

The CO₂ emissions analysis in this project is conceptually guided by the Kaya Identity framework, which explains emissions as a function of economic activity, energy intensity, and carbon intensity of the energy system. Using the available indicators within the dataset, the analysis explores how economic growth, energy efficiency, and fossil-fuel dependency influences emission patterns across African countries.


### The Kaya Identity Framework

Total national CO₂ emissions can be decomposed into four structural factors:

**CO₂ = Population × (GDP/Population) × (Energy/GDP) × (CO₂/Energy)**

Where:
- **GDP/Population** = GDP per Capita: The economic affluence and living standards of the population.
- **Energy/GDP** = Energy Intensity: The energy efficiency of the economy (how much raw energy is consumed per dollar of economic output).
- **CO₂/Energy** = carbon intensity of the energy mix (cleanliness of the grid; how much carbon is released per unit of energy generated).

This decomposition frames the two analyses below: emissions against economic development, and emissions against energy intensity.





*Note: Kaya Identity was Introduced by Japanese energy economist Yoichi Kaya in 1990, it is used to break down and analyze the exact drivers of total carbon dioxide (CO₂) emissions.*

### 12.1 CO₂ Emissions vs. Economic Development (2019)

Continental electricity access in Africa improved significantly between 2000 and 2019, rising from approximately 34% to about 54%. North Africa consistently maintained the highest and most inclusive electricity access levels throughout the period, generally exceeding 80%, while Southern Africa recorded some of the strongest overall improvements among the remaining regions. Ultimately, energy poverty across Africa remains a multidimensional challenge requiring integrated infrastructure expansion, clean energy adoption, and inclusive policy interventions.

Access to clean cooking fuels and technologies improved across all African regions between 2000 and 2019, but overall progress remained critically limited. By 2019, the continental average stood at only approximately 30%, indicating that the majority of households still depended on traditional biomass sources such as firewood, charcoal, and animal dung for cooking. North Africa consistently recorded the highest clean cooking access rates, exceeding 90%, largely due to stronger urban infrastructure development and widespread adoption of piped gas systems. Southern Africa also demonstrated steady improvement over the period, increasing from roughly 36% to 57%.

In contrast, progress in East and West Africa remained comparatively slow. These patterns highlight persistent regional disparities in access to modern household energy solutions across the continent.

The findings highlight that renewable energy share alone is insufficient for evaluating the quality of energy transition. This highlights a central sustainability challenge for the continent: expanding energy access and economic development without reproducing the high-carbon development pathways historically followed by industrialized economies.

**GDP vs CO₂:  
r ≈ 0.373, p-value ≈ 0.008;** Moderate positive relationship.
The analysis reveals a moderate positive relationship between GDP per capita and CO₂ emissions, but the relationship is uneven across countries. Countries with fossil-fuel-dependent industrial systems emit substantially more than countries with lower industrial activity or higher reliance on renewable sources. 

**Energy intensity vs CO₂:  
r ≈ 0.05, p > 0.05;** effectively no relationship. 
Energy intensity exhibits almost no meaningful relationship with emissions levels, indicating that inefficient energy use alone does not necessarily translate into higher carbon emissions across African countries. Several countries with moderate or high energy intensity still maintain relatively low emissions due to limited industrial activity and smaller-scale energy systems.

**CO₂ vs. Fossil electricity generation:  
r ≈ 0.797, p ≈ 0.0000**
Fossil-fuel electricity generation shows a much stronger and statistically significant relationship with CO₂ emissions, emphasizing that the carbon composition of the electricity system plays a more important role in determining emissions than efficiency metrics alone. Countries characterized by high emissions and low shares of low-carbon electricity typically face structural dependence on fossil-based power systems, suggesting that long-term decarbonization strategies and cleaner electricity generation are likely to be more impactful than efficiency improvements alone.


#### Summary of Findings
1. Access gap is large and uneven. Continental electricity access improved from ~25% to ~55% (2000–2019). Progress is real but the 90-pp gap between best and worst countries, and a continental median clean cooking access of just 15%, confirm that Africa remains far from SDG 7 targets.
2. Renewable share is a misleading headline metric. High shares in East and Central Africa are biomass-driven, not modern renewables. Renewable capacity per capita is the more reliable progress indicator.
3. Economic growth alone does not electrify countries. Rwanda, Kenya, and Ghana outperform their income-predicted access through deliberate electrification policy.
4. Emissions are concentrated, not continental. South Africa, Egypt, Algeria, and Nigeria dominate. Most African nations are negligible emitters globally. 
5. External financing is necessary but insufficient. Three separate correlation tests found no statistically significant link between financial flows and renewable adoption, share, or capacity. Effectiveness depends on governance, institutional capacity, and domestic co-investment not funding volume alone.
6. Regional divergence is widening, not narrowing. North Africa’s advantage over African Regions in electricity access, clean fuels, and GDP is structural. Without targeted regional investment, the energy divide might persist for decades.



## 16. Recommendations

### Infrastructure & Policy

 **i). Expand modern electricity infrastructure in underserved regions**

Large parts of Sub-Saharan Africa, particularly in West and Central Africa, still experience limited electricity access despite improvements in renewable energy generation. Expanding grid infrastructure, rural electrification programs, and decentralized energy systems such as mini-grids and solar home systems would help reduce energy poverty and improve economic productivity.

 **ii). Accelerate clean cooking fuel programs**

The analysis revealed that clean cooking fuel access remained critically low in several regions despite improvements in electricity systems. Governments and development agencies should prioritize affordable LPG distribution, electric cooking technologies, biogas systems, and public awareness programs to reduce dependence on biomass fuels such as wood and charcoal, which contribute to deforestation and indoor air pollution.

 **iii). Strengthen grid reliability and transmission systems**

Several countries continue to face weak transmission infrastructure, unstable electricity supply, and poor regional connectivity. Investment in transmission networks, grid modernization, and storage technologies is necessary to improve electricity reliability and support large-scale renewable integration.


### Renewable Transition

**i). Move beyond traditional biomass dependence**

High renewable energy shares in many African countries are still heavily influenced by traditional biomass use rather than modern renewable infrastructure. Policymakers should focus on transitioning toward cleaner and more scalable renewable systems such as solar, wind, geothermal, and modern hydropower rather than relying on unsustainable biomass consumption.

**ii). Scale utility-scale solar, wind, and geothermal deployment**

Africa possesses significant untapped renewable energy potential, particularly in solar and geothermal resources. Expanding utility-scale renewable projects can improve electricity generation capacity, reduce fossil fuel dependence, and support long-term energy security while maintaining relatively low emissions growth.

**iii). Improve regional grid integration**

Cross-border electricity integration can help balance energy supply variability and improve energy access across regions. Strengthening regional power pools and transmission interconnections would allow countries with strong renewable resources to export electricity to neighboring states with energy deficits.


### Decarbonization

**i). Reduce fossil-fuel electricity dependence in industrialized economies**

The analysis showed that high-emission countries such as South Africa and several North African economies remain heavily dependent on fossil-fuel-based electricity generation. Transitioning toward lower-carbon electricity systems is essential for reducing long-term emissions growth while maintaining industrial productivity.

**ii). Prioritize coal transition strategies in Southern Africa**

Southern Africa, particularly South Africa, remains one of the continent’s most carbon-intensive energy regions because of coal-based electricity systems. Gradual coal transition strategies involving renewable expansion, grid modernization, and industrial decarbonization policies would significantly reduce regional emissions intensity.



### Financial Development

**i). Improve accountability and efficiency of foreign financial assistance**

The relationship between foreign financial assistance and renewable energy outcomes was relatively weak, suggesting that funding alone does not guarantee successful infrastructure development. Improving project transparency, governance, and implementation efficiency would help ensure that financial inflows produce measurable sustainability outcomes.

**ii). Align funding toward measurable infrastructure outcomes**

International energy financing should increasingly prioritize projects with clear long-term impacts on electricity access, renewable generation capacity, clean cooking adoption, and grid modernization. Linking funding to measurable development indicators would improve the effectiveness of sustainability investments across the continent.
